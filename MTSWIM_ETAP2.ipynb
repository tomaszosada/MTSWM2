```python
#X = test_data.reset_index()
# classes['KLASA']
dataset = np.genfromtxt("bialaczka.csv" , delimiter=";")
dataset = dataset.astype(np.int32)
X = dataset[1:, 2:-1]
y = dataset[1:, -1].astype(int)
# print(y)
```

```python
clfs= {
    'model1Euc': KNeighborsClassifier(n_neighbors=1,metric="euclidean"),
    'model5Euc': KNeighborsClassifier(n_neighbors=5,metric="euclidean"),
    'model10Euc': KNeighborsClassifier(n_neighbors=10,metric="euclidean"),
    'model1Cheb': KNeighborsClassifier(n_neighbors=1,metric="chebyshev"),
    'model5Cheb': KNeighborsClassifier(n_neighbors=5,metric="chebyshev"),
    'model10Cheb': KNeighborsClassifier(n_neighbors=10,metric="chebyshev")
}
print(X)
print(len(X))

features= {
    '1 feature' : (X[:, [3]]),
    '2 features': (X[:, [3,16]]),
    '3 features': (X[:, [3,16, 18]]),
    '4 features': (X[:, [3,16, 18, 1]]),
    '5 features': (X[:, [3,16, 18, 1, 13]]),
    '6 features': (X[:, [3,16, 18, 1, 13, 17]]),
    '7 features': (X[:, [3,16, 18, 1, 13, 17, 11]])
}

# for a, b in enumerate(features):
#     print(features[b])
random_state = 42
n_splits = 2
n_repeats = 5
rskf = RepeatedStratifiedKFold(n_splits = n_splits, n_repeats = n_repeats, random_state = random_state)

scores = np.zeros((len(features), len(clfs),  n_splits * n_repeats))

for feature_id, feature_name in enumerate(features):
    X = features[feature_name]
    for fold_id, (train, test) in enumerate(rskf.split(X, y)):
        for clf_id, clf_name in enumerate(clfs):
            clf = (clfs[clf_name])
            clf.fit(X[train], y[train])
            y_pred = clf.predict(X[test])
            scores[feature_id, clf_id, fold_id] = accuracy_score(y[test], y_pred)
```

```python
mean = np.mean(scores, axis=1)
std = np.std(scores, axis=1)
for feature_id, feature_name in enumerate(features):
    for clf_id, clf_name in enumerate(clfs):
        print(" %s, %s: %.3f (%.2f)" % (feature_name, clf_name, mean[feature_id][clf_id], std[feature_id][clf_id]))
```

```python
alfa = 0.05
t_statistic = np.zeros((len(clfs), len(clfs)))
p_value = np.zeros((len(clfs), len(clfs)))

for x in range(len(features)):
    for i in range(len(clfs)):
        for j in range(len(clfs)):
            t_statistic[i, j], p_value[i, j] = ttest_ind(scores[x][i], scores[x][j])
    print("\n" + str(x + 1) + " features:")
    print("t-statistic:\n", t_statistic, "\n\np-value:\n", p_value)

```
